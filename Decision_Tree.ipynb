## Backward Elimination

```python
! pip install statsmodels
```

```python
import statsmodels.regression.linear_model as sm 
import numpy as np 


def backwardElimintion(X,y,sl,numVar,columns):
    
    for i in range(numVar):
        resgressor=sm.OLS(y,X).fit()
        max_p_value=max(resgressor.pvalues)
        
        if max_p_value > sl :
            for j in range(0,numVar-i):
                if resgressor.pvalues[j]==max_p_value:
                    X=np.delete(X,j,1)
                    columns=np.delete(columns,j)
        else:
            break
            
    return X,columns


features,columns=backwardElimintion(selected_features.iloc[:,:-1].values,selected_features.iloc[:,-1].values,0.05,selected_features.shape[1]-1,selected_features.columns[:-1])
```

```python
final_data=pd.DataFrame(features,columns=columns)
final_data['Price']=selected_features['Price'].values
final_data.head()
```

```python
X=final_data.iloc[:,:-1].values
y=final_data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.2) # train 80% ,test20%
```

### Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(max_depth=20,min_samples_split=3)
```

```python
model.fit(train_X,train_y)
```

### evaluate model

```python
y_pred=model.predict(test_X)
```
