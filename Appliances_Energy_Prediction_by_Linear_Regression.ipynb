## Handle Categorical Values

```python
y=data['Appliances']
data=pd.get_dummies(data.iloc[:,:-1],drop_first=True)
```

```python
data['Appliances']=y.values
data.head()
```

```python
len(data)
```

```python
data.columns
```

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2) # train 80% ,test20%
```

```python
from sklearn.preprocessing import StandardScaler 
# Scale the data
scaler = StandardScaler() 
X_train_sc=scaler.fit_transform(X_train)
X_test_sc=scaler.transform(X_test)
y_train=y_train.reshape(-1,1)
y_test=y_test.reshape(-1,1)
y_train_sc=scaler.fit_transform(y_train)
y_test_sc=scaler.transform(y_test)
```

### Decision tree

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
model.fit(X_train_sc,y_train_sc)
```

### evaluate model

```python
y_pred=model.predict(X_test_sc)
```

```python
from sklearn import metrics
mse=metrics.mean_squared_error(y_pred,y_test_sc)
print(mse)
```
